1. # What are data structures, and why are they important?
   - Data structures are specialized formats for organizing storing, and  accessing collections of data.They areimportant because they
   Act as containers to hold data,
   Simplify data manipulation,
   Optimize searching and storing operations,
   Conserve memory usage.

2. # Explain the difference between mutable and immutable data types with examples.
   - Mutable data types: These are data types whose values can be changed after creation. like list, dictionary.
   Immutable data types: These are data types whose values cannot be changed after creation. like tuple, string.

3. # What are the main differences between lists and tuples in Python?
   - Lists are mutable, use square brackets [""], slower compared to tuples, and consume more memory.Tuples are immutable, use parentheses (), faster compared to lists, and support fewer functions like count() and index().

4. # Describe how dictionaries store data.
   - A dictionary in Python is a data structure that stores data in the form of key-value pairs. Every key is like a special name that helps you find its value, and the data associated with that key is called the value.

5. # Why might you use a set instead of a list in Python?
   - Because sets are an unordered collection of unique elements, and they are useful for finding common elements or removing duplicates.

6. # What is a string in Python, and how is it different from a list?
   - A string in Python is a sequence of characters used to represent text and is immutable, meaning its contents cannot be changed once created. A list, on the other hand, can store multiple data types, is mutable, and allows modification of its elements, while both support indexing and slicing.

7. # How do tuples ensure data integrity in Python?
   - Tuples ensure data integrity in Python by being immutable, meaning their values cannot be changed once created.

8. # What is a hash table, and how does it relate to dictionaries in Python?
   - A hash table is a data structure that stores values using keys, with a hash function mapping each key to an index for fast access. A Python dictionary is an implementation of a hash table, where data is stored as key value pairs.

9. # Can lists contain different data types in Python?
   - Yes, a list in Python can contain elements of different data types.
like int, float, complex, str, tuple, another list, range, dict, set, frozenset, bool, etc

10. # Explain why strings are immutable in Python?
    - Strings are immutable because immutability makes them memory-efficient, reliable as dictionary keys, thread-safe, and consistent to use.

11. # What advantages do dictionaries offer over lists for certain tasks?
    - Dictionaries offer faster lookups, store data as key value pairs, ensure unique keys, and allow direct access using keys instead of indexes.

12. # Describe a scenario where using a tuple would be preferable over a list?
    - When you create employee IDs, memberships, or permanent details, you must use tuples because they cannot be modified or changed. Like a list

13. # How do sets handle duplicate values in Python?
    - The purpose of a set is to store unique values, so if you add a duplicate, Python automatically keeps only one copy and removes the rest.

14. # How does the “in” keyword work differently for lists and dictionaries.
    - The "in" keyword checks if a value/element exists in the list.The "in" keyword checks if a key exists in the dictionary (not the value).

15. # Can you modify the elements of a tuple? Explain why or why not?
    - A tuple is immutable, which means once it is created, its elements cannot be changed, added, or removed.If a tuple contains a mutable object like a list, the contents of that object can be modified, but the tuples structure the positions of elements itself still cannot change.

16. # What is a nested dictionary, and give an example of its use case
    - A nested dictionary means storing one dictionary inside another. It is useful for organizing complex data in a structured way, like storing student details name, marks, age subject-wise.

17. # Describe the time complexity of accessing elements in a dictionary.
    - Because dictionaries use hashing, so Python can directly jump to the memory location of the key.

18. # In what situations are lists preferred over dictionaries
    - Lists are preferred when we need to store data in an ordered sequence, access elements by index, or allow duplicate values. They are useful for simple collections like numbers, names, or items.

19. # Why are dictionaries considered unordered, and how does that affect data retrieval?
    - Dictionaries are considered unordered because they store data based on hashing, not in a fixed sequence.Effect on retrieval you can access values only by their keys, not by index like a list.

20. # Explain the difference between a list and a dictionary in terms of data retrieval.
    - List is an ordered collection. Data is retrieved from index number. Dictionary is an unordered collection. Data is retrieved from unique key.
